```python
import matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
ddir='/Users/yunyao/Dropbox/LeuT/1D1_AA_dis'
```

```python
df=pd.DataFrame(columns=['res','res_1L', 'No.of_res','perc'])
```

```python
with open (ddir, 'r' ) as item:
    for line in item:
        newline=line.rstrip('\n').rstrip('\\').rstrip('\t').rstrip('\ ').split()
        #a=print(newline)
        #a_series= pd.Series(newline, index=df.columns)
        df=df.append({'res':newline[0], 'res_1L':newline[1].strip('()'),'No.of_res':int(newline[2]), 'perc':float(newline[3].strip('%'))}, ignore_index=True)
#newline[0] 
#df.append({'res':newline[0], 'res_1L':newline[1],'No.of_res':int(newline[2]), 'perc':newline[3].strip('%')}, ignore_index=True)

        
        
        
        
        
```

```python
width=0.65
x=np.arange(len(df['res']))
fig, ax = plt.subplots()
labels=df['res_1L']
dist=ax.bar(x, df['perc'], width)
ax.set_ylabel('Amino Acid percentage %')
ax.set_title('Amino Acids Distribution in 1D1')
ax.set_xticks(x)
ax.set_ylim((0,20))
ax.set_xticklabels(labels)

def autolabel(rects,labels):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for (rect,  label) in zip(rects, labels):
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        #print(label)
        
        ax.annotate('{}'.format(label), 
                   xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 18),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', bbox=dict(boxstyle='round,pad=0.2', fc='yellow', alpha=0.3))
#rects=df['No.of_res']
autolabel(dist,df['No.of_res'] )
fig.tight_layout()
fig.savefig('1D1_AA_distribution.png', format="png",dpi=1000)

#ax.legend()
```
